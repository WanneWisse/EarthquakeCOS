```python
from TFIDF import get_tf_idf_matrix
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import helpers
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df = helpers.load_data_exp_1('preprocessed_tweets.csv')
vectorised_df = get_tf_idf_matrix(df)

original_data = df.copy()
subset = vectorised_df.copy().sample(n=5000)
```

```python
# Elbow method
k_values = range(0, 50)

# Calculate the within-cluster sum of squares (WCSS) for each K value
wcss = []
for k in k_values:
    kmeans = KMeans(n_clusters=k, n_init=10)
    kmeans.fit(subset)
    wcss.append(kmeans.inertia_)

# Plot the WCSS values against K values
plt.plot(k_values, wcss, 'bx-')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.title('Elbow Method')
plt.show()
```

```python
print(subset.shape)
k = 37  # Number of clusters
kmeans = KMeans(n_clusters=k)

kmeans.fit(subset)
labels = kmeans.labels_

centroids = kmeans.cluster_centers_
unique_labels, counts = np.unique(labels, return_counts=True)

# Print the label distribution
for label, count in zip(unique_labels, counts):
    print(f"Label {label}: {count} occurrences")


original_data.loc[:, 'label'] = labels
subset_label_2 = original_data[original_data['label'] == 2]
subset_label_1 = original_data[original_data['label'] == 1]
subset_label_0 = original_data[original_data['label'] == 0]
subset_label_2.head(5)
subset_label_1.head(5)
```
